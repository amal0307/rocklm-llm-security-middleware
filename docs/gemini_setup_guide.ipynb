# Fixing Google Generative AI Module Errors

This guide will help you resolve the `module 'google.generativeai' has no attribute 'GenerativeModel'` error and set up the Google Generative AI SDK correctly.

## 1. Environment Setup and Verification

First, let's check our Python environment and set up a clean virtual environment. Based on reported issues, Python 3.10 is recommended for optimal compatibility.

```python
# Check Python version
import sys
print(f"Python version: {sys.version}")

# Check if we're in a virtual environment
def check_venv():
    return sys.prefix != sys.base_prefix

print(f"\nRunning in virtual environment: {check_venv()}")

# List installed packages
import pkg_resources
print("\nInstalled packages:")
for pkg in pkg_resources.working_set:
    if "google" in pkg.key or "genai" in pkg.key:
        print(f"{pkg.key}=={pkg.version}")
```

## 2. Package Installation and Version Check

Let's install the correct version of the Google Generative AI package. The recommended approach is to:
1. Upgrade pip
2. Uninstall any existing google-generativeai package
3. Install the latest version

**Note:** If you're using Python 3.11+, you might need to downgrade to Python 3.10 as there are known compatibility issues with newer Python versions.

```python
# Run these commands in your terminal:
# %pip install --upgrade pip
# %pip uninstall -y google-generativeai
# %pip install --upgrade google-generativeai

import google.generativeai as genai
print(f"google-generativeai version: {genai.__version__}")

# Check if GenerativeModel exists
print("\nChecking for GenerativeModel attribute:")
print("GenerativeModel exists:", hasattr(genai, "GenerativeModel"))
```

## 3. Working Implementation

Here's the correct way to use the Google Generative AI SDK with Python 3.10 and the latest package version:

```python
import os
import google.generativeai as genai

# Configure the API
genai.configure(api_key="YOUR_API_KEY")  # Replace with your actual API key

# Create a model instance
model = genai.GenerativeModel("models/gemini-1.5-pro-002")

# Test the model
try:
    response = model.generate_content("Hello! How are you?")
    print("Model response:", response.text)
    print("\nTest successful! The API is working correctly.")
except Exception as e:
    print("Error:", str(e))
```

## 4. Common Issues and Solutions

1. **Wrong Python Version**
   - Problem: Python 3.11+ might have compatibility issues
   - Solution: Use Python 3.10 (recommended) or Python 3.9

2. **Old Package Version**
   - Problem: Versions below 0.3.0 don't have GenerativeModel
   - Solution: Upgrade to latest version (0.5.2+)

3. **Environment Issues**
   - Problem: Package installed in wrong environment
   - Solution: Use %pip install in Jupyter or verify virtual environment

4. **Import Issues**
   - Problem: Wrong import statement or initialization
   - Solution: Use `import google.generativeai as genai` and initialize with `genai.configure()`

5. **Runtime Issues**
   - Problem: Notebook kernel not restarted after installation
   - Solution: Restart kernel after package installation/upgrade
